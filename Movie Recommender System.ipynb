Now converting the movie genres into a list of items

```python
import ast
def convertToList(lst_obj):
  lst = list()
  for item in ast.literal_eval(lst_obj):    # because the list is initially a string object
    lst.append(item['name'])
  return lst
```

```python
movies['genres'] = movies['genres'].apply(convertToList)
movies['keywords'] = movies['keywords'].apply(convertToList)
```

```python
def convertToList2(lst_obj):
  lst = list()
  counter = 0
  for item in ast.literal_eval(lst_obj):
    if counter != 3:
      lst.append(item['name'])
      counter += 1
    else:
      break
  return lst
```

```python
movies['cast'] = movies['cast'].apply(convertToList2)
```

```python
movies.head(2)
```

```python
def fetchDierctor(lst_obj):
  lst = list()
  for item in ast.literal_eval(lst_obj):
    if item['job'] == 'Director':
      lst.append(item['name'])
      break
  return lst
```

```python
movies['crew'] = movies['crew'].apply(fetchDierctor)
```

```python
movies.head(2)
```

### **Applying the vectorization techniques to find out the similar tags between the movies**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

Insights:
- there are multiple vocabs with similar names, ie action and actions, two similar words but occuring as two different vectors

so we shal apply stemming in this case

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
print(cv.get_feature_names_out())
```

**Now finding the cosine similarity between the vectors to identify the most similar vectors**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
print(similarity[0])
```

```python
new_df.info()
```

```python
def recommend(movie):   # recommends top 5 similar movies
  try:
    movie_index = new_df[new_df['title'] == movie.title()].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for movie in movies_list:
      print(new_df.iloc[movie[0]].title)
  except IndexError as e:
    print(f"Sorry \'{movie}\' movie not found!\nMake sure that the spelling of the movie name is correct\n")
```

```python
recommend('batman begins')
```

```python
recommend('avatar')
```

```python
recommend('avatra')
```
